### Evaluation Metrics - F1 Score and Accuracy Score using Gradient Descent

### Evaluation Metrics - F1 Score and Accuracy Score using Maximum Likelihood

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('NYPD_Hate_Crimes.csv')

features = ['Bias Motive Description', 'Month Number', 'Patrol Borough Name']
target = data['Arrest Id'].notnull().astype(int)

encoded_data = pd.get_dummies(data[features])

X_train, X_test, y_train, y_test = train_test_split(encoded_data, target, test_size=0.2, random_state=42)

lr = LogisticRegression(solver='lbfgs',max_iter=1000)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print('Accuracy Maximum Likelihood: {:.7f}'.format(accuracy))
print('F1 score Maximum Likelihood: {:.7f}'.format(f1))
```
